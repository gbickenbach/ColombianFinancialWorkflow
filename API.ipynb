```python
#A PARTIR DE LA SEGUNDA VEZ CORRER ESTE CODIGO, TIEMPO ESTIMADO 4 MINUTOS
data = pd.read_csv('drive/MyDrive/Sophomore Year/Asobancaria/datapi.csv', engine ='pyarrow', usecols=['sexo'])
data = data.drop(columns=[ 'Unnamed: 0'])
```

```python
pd.set_option("display.max_columns", None)
original = data

def organizar(data, categoria1, categoria2, orden):
  if(orden == 'ascendiente' or orden == 'Ascendiente'):
    data = data.sort_values(by=[categoria1, categoria2])
    return(data)
  elif(orden == 'descendiente' or orden == 'Descendiente'):
    data = data.sort_values(by=[categoria1, categoria2], ascending=False)
    return(data)
def buscar (data, categoria, operador, especificador):
  if(operador == '='):
    data = data.loc[data[categoria] == especificador]
  elif(operador == '>'):
    data = data.loc[data[categoria] > especificador]
    data
  elif(operador == '<'):
    data = data.loc[data[categoria] < especificador]
    data
  elif(operador == '>='):
    data = data.loc[data[categoria] >= especificador]
    data
  elif(operador == '<='):
    data = data.loc[data[categoria] <= especificador]
    data
  return(data)
def guardar(data, nombre):
  data.to_csv('drive/MyDrive/Sophomore Year/Asobancaria/'+ nombre +'.csv')
def abrir(nombre):
  data = pd.read_csv('drive/MyDrive/Sophomore Year/Asobancaria/'+ nombre +'.csv')

```
